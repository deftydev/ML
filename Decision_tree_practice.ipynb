```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix , roc_curve, roc_auc_score
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv('winequality-red.csv')
```

```python
df.head()
```

```python
ProfileReport(df)
```

```python
dt_model = DecisionTreeClassifier()
```

```python
x= df.drop('quality', axis=1)
```

```python
y= df['quality']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.40,random_state=400)
```

```python
dt_model.fit(x_train,y_train)
```

```python
dt_model.predict(x_test)
```

```python
dt_model.score(x_test,y_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,20))
tree.plot_tree(dt_model,filled=True)
```

```python
df1 = df.head(50)
x1 = df1.drop(columns='quality')
y1 = df1.quality
dt_model1 = DecisionTreeClassifier()
dt_model1.fit(x1,y1)
plt.figure(figsize=(20,20))
tree.plot_tree(dt_model1,filled=True, class_names=[str(i) for i in set(y1)] ,feature_names=x1.columns)
plt.savefig('dt_model_1')
```

```python
df11 = df.head(500)
x11 = df11.drop(columns='quality')
y11 = df11.quality
```

# But we should perform pruning so we should determine ccp_alpha value for our model , so we can define that using ccp_alpha values

```python
path= dt_model1.cost_complexity_pruning_path(x11, y11)
ccp_alpha = path.ccp_alphas
print(ccp_alpha)
```

```python
dt_ccp = []
for i in ccp_alpha:
    dt_new_model = DecisionTreeClassifier(ccp_alpha= i)
    dt_new_model.fit(x11,y11)
    dt_ccp.append(dt_new_model)
```

```python
for j in dt_ccp:
    print(j.score(x_test,y_test))
```

```python
train_score = [i.score(x11,y11) for i in dt_ccp]
```

```python
test_score = [i.score(x_test ,y_test) for i in dt_ccp]
```

```python
fig, ax = plt.subplots() ##from this we can easily found that for which value of ccp_alpha train and test accuracy is equal
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alpha, train_score, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alpha, test_score, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
```

# so from above figure we can say that for alpha vaue between 0.01 to 0.02 we are getting ame accuracy with training and test dataset
so we can use that value to build our model

```python
dt_model_ccp = DecisionTreeClassifier(random_state=0 , ccp_alpha=.014)
dt_model_ccp.fit(x1,y1)
```

```python
dt_model_ccp.score(x11,y11)
```

```python
dt_model_ccp.score(x_test,y_test)
```
